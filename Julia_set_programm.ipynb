```python
import numpy as np
import matplotlib.pyplot as plt
import random

def julia_set(z, c):
    """    """
    n = 100
    while abs(z) < 10 and n > 0:
        z = z**2 + c
        n -= 1
    return n

#   c
c = complex(-0.01743,0.78677)
print(c)

#      
x_min, x_max, y_min, y_max = -1.5, 1.5, -1.5, 1.5
resolution = 1000
x = np.linspace(x_min, x_max, resolution)
y = np.linspace(y_min, y_max, resolution)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

#   
J = np.frompyfunc(julia_set, 2, 1)(Z, c).astype(np.float)

#   
plt.imshow(J, cmap='magma', extent=(x_min, x_max, y_min, y_max))
plt.title(f"   c={c}")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.show()
```
