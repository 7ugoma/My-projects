```python
import matplotlib.pyplot as plt
import math

# Функция для рисования Дерева Пифагора
def draw_pifagoras_tree(x, y, angle, depth, length, width):
    if depth == 0:
        return

    # Вычисляем координаты конечной точки ветви
    x_end = x + length * math.cos(math.radians(angle))
    y_end = y + length * math.sin(math.radians(angle))

    # Рисуем ветвь
    plt.plot([x, x_end], [y, y_end], color='brown', linewidth=width)

    # Вызываем рекурсивно функцию для рисования правой ветви
    draw_pifagoras_tree(x_end, y_end, angle - 30, depth - 1, length * 0.6, width * 0.7)

    # Вызываем рекурсивно функцию для рисования левой ветви
    draw_pifagoras_tree(x_end, y_end, angle + 30, depth - 1, length * 0.6, width * 0.7)

# Создаем фигуру и оси
fig, ax = plt.subplots()

# Устанавливаем размер окна
plt.figure(figsize=(8, 8))

# Устанавливаем пределы осей
ax.set_xlim([-100, 100])
ax.set_ylim([0, 200])

# Вызываем функцию для рисования Дерева Пифагора
draw_pifagoras_tree(0, 0, 90, 11, 100, 10)

# Устанавливаем заголовок и отключаем отображение осей
ax.set_title('Дерево Пифагора')
ax.set_axis_off()

# Отображаем фрактал
plt.show()
```
