```python
import matplotlib.pyplot as plt
import random
```

```python
# Исходные параметры для преобразований Папоротника Барнсли
# в виде списка кортежей вида (коэффициенты a, b, c, d, f)
TRANSFORMS = [([0.75, 0.04, -0.04, 0.88, 0, 1.6], 0.85),
             ([-0.15, 0.28, 0.26, 0.24, 0, 0.44], 0.07),
             ([0.17, -0.27, 0.23, 0.24, 0, 1.6], 0.07),
             ([0, 0, 0, 0.16, 0, 0], 0.01)]

# Исходная точка
START_POINT = [0, 0]

# Количество итераций
NUM_ITERATIONS = 50000

# Списки для хранения координат точек
x = []
y = []

# Функция для выбора случайного преобразования
def choose_transform(transforms, probabilities):
    """Выбирает случайное преобразование на основе заданных вероятностей"""
    return random.choices(transforms, probabilities)[0]

# Начальная точка
point = START_POINT

# Генерация фрактала
for _ in range(NUM_ITERATIONS):
    # Выбор случайного преобразования и его вероятности
    transform, probability = choose_transform(TRANSFORMS, [p for _, p in TRANSFORMS])
    # Применение выбранного преобразования
    point = [transform[0] * point[0] + transform[1] * point[1] + transform[4],
             transform[2] * point[0] + transform[3] * point[1] + transform[5]]
    # Добавление координат точки в списки
    x.append(point[0])
    y.append(point[1])

```

```python
# Построение графика
plt.scatter(x, y, s=0.1, c='green', edgecolors='none')
plt.title('Папоротник Барнсли')
plt.axis('off')
plt.show()
```
